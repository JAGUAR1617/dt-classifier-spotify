```
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz
import pydotplus
from scipy import misc

import io
%matplotlib inline
```

```
c = DecisionTreeClassifier(min_samples_split=100)  # min_sample_split = 2, means very dense tree ever high risk of overfitting
```

```
features = ["danceability", "duration_ms",  "loudness",  "speechiness",  "valence",  "energy", "acousticness",  "key",  "instrumentalness"]
```

```
features
```

```
X_train = train[features]
y_train = train["target"]

X_test = train[features]
y_test = train["target"]
```

```
y_test
```

```
dt = c.fit(X_train, y_train)
```

```
import imageio
```

```
def show_tree(tree, features, path):
  f = io.StringIO()
  export_graphviz(tree, out_file=f, feature_names=features)
  pydotplus.graph_from_dot_data(f.getvalue()).write_png(path)
  img = imageio.imread(path)
  plt.rcParams["figure.figsize"] = (20, 20)
  plt.imshow(img)
```

```
show_tree(dt, features, 'de_tree_01.png')  #any file name to wrrite
```

```
y_pred = c.predict(X_test)
```

```
y_pred
```

```
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_pred)*100
```

```
print("accuracy using dt: ", round(score, 1), "%")
```
